```python
import requests
import json

def extract_tokaopedia_reviews(url_target, header, product_id, page=1, limit=10, sort_by="informative_score desc"):
    """
    Extracts product review data from Tokopedia's productReviewList.

    Args:
        url_target: The URL of the Tokopedia GraphQL endpoint.
        header: The HTTP headers to use for the request.
        product_id: The ID of the product for which to fetch reviews.
        page: The page number (default: 1).
        limit: The number of reviews per page (default: 10).
        sort_by: The field to sort reviews by (default: "informative_score desc").

    Returns:
        A list of extracted review dictionaries, or None if an error occurs.
    """

    variables = {
        "productID": product_id,
        "page": page,
        "limit": limit,
        "sortBy": sort_by,
        "filterBy": "",  # Optional filter, currently empty
    }

    query = f"""
    {{
      "operationName": "productReviewList",
      "variables": {json.dumps(variables)},
      "query": "query productReviewList($productID: String!, $page: Int!, $limit: Int!, $sortBy: String, $filterBy: String) {{\\n"
                f"  productrevGetProductReviewList(productID: $productID, page: $page, limit: $limit, sortBy: $sortBy, filterBy: $filterBy) {{\\n"
                f"    productID\\n"
                f"    list {{\\n"
                f"      # ... desired review fields ... (e.g., id, variantName, message, rating, timestamps, ...) \\n"
                f"    }}\\n"
                f"    shop {{\\n"
                f"      shopID\\n"
                f"      name\\n"
                f"      url\\n"
                f"      image\\n"
                f"      __typename\\n"
                f"    }}\\n"
                f"    hasNext\\n"
                f"    totalReviews\\n"
                f"    __typename\\n"
                f"  }}\\n"
                f"}}\\n"
    }}
    """

    try:
        response = requests.post(url_target, headers=header, data=query)
        response.raise_for_status()  # Raise an exception for non-200 status codes
        data = response.json()[0]['data']['productrevGetProductReviewList']

        reviews = [
            {
                # Extract desired review data fields
            }
            for review in data['list']
        ]

        return reviews
    except requests.exceptions.RequestException as e:
        print(f"Error occurred: {e}")
        return None

# Example usage:
url_target = "https://your-tokopedia-graphql-endpoint"
header = {"Content-Type": "application/json"}
product_id = "1234567890"

reviews = extract_tokaopedia_reviews(url_target, header, product_id)

if reviews:
    # Process extracted review data
    print(f"Number of reviews: {len(reviews)}")
else:
    print("Error fetching reviews")
```
